Advantage of implicit representations

```julia
sizeof(bigarrary)
```

this is the amount of memory used to store the datea in bytes

```julia
bigarray_conrete = collect(bigarrary);
sizeof(bigarray_conrete)
```

```julia
sizeof(bigarray_conrete) / sizeof(bigarrary)
```

2500 times as much memory used for the explicit represetnation

```julia
using LinearAlgebra
```

Solve $Ax =v$ for $v$:

```julia
x = A\v
```

```julia
A*x
```

```julia
v
```

Note a slight floating point error and come to terms with it.

Matrix inverse:

```julia
inv(A)
```

If your goal is to solve $Ax=b$, use `x = A\b`, **not** `x = inv(A)*b`

```julia
det(A)
```

```julia
eigvals(A) #igenvalues of A
```

```julia
eigvecs(A) #eigenvectors of A
```

# Visualizations
Load `Plots`:

```julia
using Plots
```

Plot the above $f(x)$ for $x \in [-2,2]$.  We create $(x_i, y_i)$ data:

```julia
length(x)
```

```julia
size(x)
```

```julia
x = LinRange(-2,2, 100);
y = f.(x) # evaluate f at each point in x via . command, and store in y
```

```julia
plot(x,y)
```

```julia
plot(x,y, lw = 2)
xlabel!("x")
ylabel!("f(x)")
title!("Plot of f(x) = x^2 + 2x + 1")
```

```julia
scatter(x,y, lw = 2)
xlabel!("x")
ylabel!("f(x)")
title!("Plot of f(x) = x^2 + 2x + 1")
```
